# **Region wise Data**

```python
dt.head()
```

```python
#filtering and sorting data
dtn = dt[(dt.Regions.isin(["NR"]))]   #NR = North Region
dtw = dt[dt['Regions']=='WR']         #WR = West Region
dts = dt[dt['Regions']=='SR']         #SR = South region
dtne = dt[dt['Regions']=='NER']       #NER = North East Region
dte = dt[dt['Regions']=='ER']         #ER = East Region
```

```python
# cummulative region wise analysis
dtw1 = dtw.groupby(['Dates'])['Usage'].sum().reset_index()
dte1 = dte.groupby(['Dates'])['Usage'].sum().reset_index()
dtn1 = dte.groupby(['Dates'])['Usage'].sum().reset_index()
dts1 = dts.groupby(['Dates'])['Usage'].sum().reset_index()
dtne1 = dtne.groupby(['Dates'])['Usage'].sum().reset_index()
```

```python
#merging Data

dta = pd.merge(dtw1, dte1, on= 'Dates' )
dta = dta.rename( columns={'Usage_x':'WR', 'Usage_y':'ER'})

dtb = pd.merge(dta, dtn1, on = 'Dates')
dtb = dtb.rename( columns={'Usage':'NR'})

dtc = pd.merge(dtb, dts1, on='Dates')
dtc = dtc.rename( columns={'Usage': 'SR'})

dt1 = pd.merge(dtc, dtne1, on='Dates')
dt1 = dt1.rename( columns={'Usage':'NER'})

dt1.head()
```

```python
#graph
figdt1=go.Figure()
figdt1.add_trace(go.Scatter(x=dt1['Dates'].values, y=dt1['WR'].values,
                         mode='lines',
                         name='WR'))
figdt1.add_trace(go.Scatter(x=dt1['Dates'].values, y=dt1['ER'].values,
                         mode='lines',
                         name='ER'))
figdt1.add_trace(go.Scatter(x=dt1['Dates'].values, y=dt1['NR'].values,
                         mode='lines',
                         name='NR'))
figdt1.add_trace(go.Scatter(x=dt1['Dates'].values, y=dt1['SR'].values,
                         mode='lines',
                         name='SR'))
figdt1.add_trace(go.Scatter(x=dt1['Dates'].values, y=dt1['NER'].values,
                         mode='lines',
                         name='NER'))

figdt1.update_layout(title='Region wise compare')
figdt1.show()
```
